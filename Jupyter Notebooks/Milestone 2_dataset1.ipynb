# Dataset 1: USDA Food Access Research Atlas

## Overview

This dataset is sourced from the USDA Food Access Research Atlas and provides county-level indicators related to food access across the United States. The dataset includes measures of population distance to grocery stores, income-related food access barriers, and low-access population counts.

## Research Relevance

Food access is a critical dimension of food security. Counties with limited access to grocery stores often rely on convenience retail, which may provide limited access to nutritious food. This dataset enables analysis of geographic disparities in food access across the United States.

## Data Source

USDA Food Access Research Atlas

## Dataset 1: Food Access Research Atlas (Raw Variables)

The dataset contains census tract–level indicators related to food access, socioeconomic conditions, and demographic characteristics. The variables can be grouped into several categories.

### 1. Geographic Identifiers

| Column | Description |
|------|-------------|
| CensusTract | Unique census tract identifier |
| State | U.S. state name |
| County | County name |
| Urban | Indicator identifying whether the tract is urban or rural |

These variables allow spatial identification and enable aggregation to county-level datasets.

---

### 2. Population and Housing Characteristics

| Column | Description |
|------|-------------|
| Pop2010 | Total population in the tract (2010 Census) |
| OHU2010 | Occupied housing units |
| GroupQuartersFlag | Indicates presence of group quarters (e.g., dorms, prisons) |
| NUMGQTRS | Number of people living in group quarters |
| PCTGQTRS | Percentage of population in group quarters |

These variables provide context for population structure and housing conditions.

---

### 3. Low-Income / Low-Access Indicators (LILA)

| Column | Description |
|------|-------------|
| LILATracts_1And10 | Low-income and low-access tract (1 mile urban / 10 miles rural threshold) |
| LILATracts_halfAnd10 | Low-income and low-access tract using 0.5 mile urban threshold |
| LILATracts_1And20 | Low-income and low-access tract using 20 mile rural threshold |
| LILATracts_Vehicle | Low-income and low-access tract considering vehicle availability |
| LowIncomeTracts | Indicator identifying low-income census tracts |

These indicators classify areas that meet USDA definitions of **food deserts**.

---

### 4. Income and Poverty Indicators

| Column | Description |
|------|-------------|
| PovertyRate | Percentage of individuals living below the poverty line |
| MedianFamilyIncome | Median household income in the census tract |

These variables capture the **economic conditions influencing food access**.

---

### 5. Low Access Population Counts

| Column | Description |
|------|-------------|
| LAPOP1_10 | Population living far from grocery stores (1 mile urban / 10 miles rural) |
| LAPOP05_10 | Population living far from grocery stores (0.5 mile urban threshold) |
| LAPOP1_20 | Population living far from grocery stores using 20 mile rural threshold |
| lapophalf | Population living more than 0.5 miles from a grocery store |
| lapop1 | Population living more than 1 mile from a grocery store |
| lapop10 | Population living more than 10 miles from a grocery store |
| lapop20 | Population living more than 20 miles from a grocery store |

These variables measure **physical food access barriers**.

---

### 6. Low-Income Population with Low Access

| Column | Description |
|------|-------------|
| LALOWI1_10 | Low-income population living far from grocery stores |
| LALOWI05_10 | Low-income population living far from grocery stores (0.5 mile threshold) |
| LALOWI1_20 | Low-income population living far from grocery stores (20 mile threshold) |
| lalowi1 | Low-income population with limited access (1 mile threshold) |
| lalowi10 | Low-income population with limited access (10 mile threshold) |
| lalowi20 | Low-income population with limited access (20 mile threshold) |

These indicators capture **combined economic and geographic barriers to food access**.

---

### 7. Demographic Breakdown of Low Access Populations

The dataset provides demographic breakdowns of populations experiencing limited food access.

| Column Pattern | Description |
|---------------|-------------|
| lakids* | Children with limited food access |
| laseniors* | Seniors with limited food access |
| lawhite* | White population with limited food access |
| lablack* | Black population with limited food access |
| laasian* | Asian population with limited food access |
| lahisp* | Hispanic population with limited food access |
| lahunv* | Households without vehicles with limited food access |
| lasnap* | SNAP recipients with limited food access |

Suffixes such as `half`, `1`, `10`, and `20` represent different distance thresholds from grocery stores.

---

### 8. Share Variables

Many variables include a **"share" suffix**, which represents proportions rather than raw population counts.

| Column Pattern | Description |
|---------------|-------------|
| lapop*share | Share of population with limited food access |
| lalowi*share | Share of low-income population with limited access |
| lakids*share | Share of children with limited access |
| laseniors*share | Share of seniors with limited access |

These variables standardize measures across census tracts with different population sizes.

---

### 9. Total Population Demographics

| Column | Description |
|------|-------------|
| TractLOWI | Total low-income population in the tract |
| TractKids | Total children population |
| TractSeniors | Total senior population |
| TractWhite | Total White population |
| TractBlack | Total Black population |
| TractAsian | Total Asian population |
| TractHispanic | Total Hispanic population |
| TractSNAP | Total SNAP participants |

These variables provide baseline demographic context for the tract.

---

### Summary

The dataset contains detailed indicators capturing:

• geographic food access barriers  
• socioeconomic vulnerability  
• demographic patterns of food insecurity  

These variables enable analysis of **food access disparities and structural drivers of food insecurity across communities.**



Remove all columns ending in `"20"` (20-mile threshold variants) except `lapop20` and `lalowi20`, which are retained for supplementary analysis. The 10-mile threshold is the primary metric used throughout this project; the 20-mile columns add redundancy without additional insight at the county level.

```python
core_20 = ["lapop20", "lalowi20"]

drop_20 = [
    c for c in food_access.columns
    if c.endswith("20") and c not in core_20
]

fa_step1 = food_access.drop(columns=drop_20)

print("Shape after 20-mile cleanup:", fa_step1.shape)
```

Remove all columns ending in `"share"`. These are ratios already present in the raw data, but they were computed at the tract level against potentially inconsistent denominators. All shares will be recomputed from scratch after cleaning to ensure consistency.

```python
share_cols = [c for c in fa_step1.columns if c.endswith("share")]

fa_step2 = fa_step1.drop(columns=share_cols)

print("Shape after removing shares:", fa_step2.shape)
```

```python
"Pop2010" in fa_step2.columns
```

Zero-pad the `CensusTract` field to exactly 11 digits (the standard format for Census tract FIPS codes: 2-digit state + 3-digit county + 6-digit tract). Derive a 5-digit `CountyFIPS` field by slicing the first five characters — this will be the join key when merging with other county-level datasets.

```python
fa_step2["CensusTract"] = fa_step2["CensusTract"].astype(str).str.zfill(11)
fa_step2["CountyFIPS"] = fa_step2["CensusTract"].str[:5]
```

```python
fa_step2[['CensusTract','CountyFIPS']].head()
```

```python
numeric_cols = fa_step2.select_dtypes(include=['float64','int64']).columns

fa_step2[numeric_cols] = fa_step2[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
fa_step2.columns.tolist()
```

```python
fa_step2.dtypes.value_counts()
```

Filter out census tracts with `Pop2010 == 0`. These are non-residential tracts (e.g., military bases, uninhabited land) where per-capita ratios would be undefined or infinite. Print the number of rows removed for transparency.

```python
fa_step3 = fa_step2[fa_step2["Pop2010"] > 0].copy()

print("Removed zero-pop tracts:",
      fa_step2.shape[0] - fa_step3.shape[0])
```

Count rows where the low-access population count exceeds the total population — a logical impossibility. This surfaces a data quality issue present in the raw USDA file, caused by rounding in their estimation methodology.

```python
fa_step3[fa_step3["lapop10"] > fa_step3["Pop2010"]].shape
```

```python
invalid = fa_step3[fa_step3["lapop10"] > fa_step3["Pop2010"]]
```


Add `excess` (absolute overshoot) and `excess_ratio` (relative overshoot) columns to the invalid subset. Run `.describe()` on these to confirm that excesses are tiny in magnitude — consistent with floating-point rounding, not a systematic data error.

```python
invalid["excess"] = invalid["lapop10"] - invalid["Pop2010"]

invalid["excess_ratio"] = invalid["lapop10"] / invalid["Pop2010"]

invalid[["Pop2010","lapop10","excess","excess_ratio"]].describe()
```

```python
# Round all low-access counts to 6 decimals
cols_to_round = ["lapop10","lalowi10","lapop20","lalowi20"]

fa_step3[cols_to_round] = fa_step3[cols_to_round].round(6)

# Now enforce logical upper bound
for col in cols_to_round:
    fa_step3[col] = np.minimum(fa_step3[col], fa_step3["Pop2010"])
```

```python
(fa_step3["lapop10"] > fa_step3["Pop2010"]).sum()
```

###TRACT LEVEL DATA COMPUTATION

```python
fa_step3["tract_low_access_10_share"] = (
    fa_step3["lapop10"] / fa_step3["Pop2010"]
)

fa_step3["tract_low_income_low_access_10_share"] = (
    fa_step3["lalowi10"] / fa_step3["Pop2010"]
)

fa_step3["tract_low_access_20_share"] = (
    fa_step3["lapop20"] / fa_step3["Pop2010"]
)

fa_step3["tract_low_income_low_access_20_share"] = (
    fa_step3["lalowi20"] / fa_step3["Pop2010"]
)
```

```python
def gini(array):
    array = np.sort(array)
    n = len(array)
    cumulative = np.cumsum(array)
    return (n + 1 - 2 * np.sum(cumulative) / cumulative[-1]) / n

tract_gini = gini(fa_step3["tract_low_access_10_share"].values)
county_gini = gini(fa_county["county_low_access_10_share"].values)

tract_gini, county_gini
```
